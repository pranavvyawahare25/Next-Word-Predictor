## **1.4. Methodology:**

**# 1.4.1 Data Preprocessing: **

 The dataset is loaded into a pandas DataFrame.
The 'Type' column is dropped, and the remaining features are scaled using the StandardScaler.

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt
````

````python
data = pd.read_csv("/content/wine (1).csv")
````

# **1.4.2 Exploratory Data Analysis (EDA):**
Univariate analysis can be performed by examining the distribution and summary statistics of each feature.
Bivariate analysis can be conducted by exploring the relationships between pairs of features using scatter plots or correlation analysis.

````python
# Data Preprocessing
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.drop('Type', axis=1))
````

# **1.4.3 Model Building:**


1.   Clustering is performed on the scaled data using K-means and hierarchical clustering algorithms.
2.  PCA is applied to reduce the dimensionality of the data to 3 principal components.
3. Clustering is performed again on the PCA-transformed data using K-means and hierarchical clustering algorithms.
4. The explained variance ratio of the principal components is printed to understand the variance captured by each component.
5. The clustering labels for both scaled data and PCA-transformed data are printed.








````python
# Model Building
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)
````

````python
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(scaled_data)
````

````python
pca = PCA(n_components=3)
pca_data = pca.fit_transform(scaled_data)
````

````python
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)
````

````python
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca_labels = kmeans_pca.fit_predict(pca_data)
````

````python
hierarchical_pca = AgglomerativeClustering(n_clusters=3)
hierarchical_pca_labels = hierarchical_pca.fit_predict(pca_data)
````

````python
print("K-means Clustering Labels (Scaled Data):", kmeans_labels)
print("Hierarchical Clustering Labels (Scaled Data):", hierarchical_labels)
print("K-means Clustering Labels (PCA):", kmeans_pca_labels)
print("Hierarchical Clustering Labels (PCA):", hierarchical_pca_labels)
````

# **1.4.4 Visualization:**
   - A scatter plot of the first two principal components is created, with each data point colored according to its assigned cluster.

````python
# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=kmeans_pca_labels, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering (PCA)')
plt.colorbar(label='Cluster')
plt.show()
````
